```python
import pandas as pd
from collections import Counter

data = {
    "Swim": [
        "Fast",
        "Fast",
        "Slow",
        "Fast",
        "No",
        "No",
        "No",
        "Slow",
        "Slow",
        "Slow",
        "No",
        "Fast",
    ],
    "Fly": [
        "No",
        "No",
        "No",
        "No",
        "Short",
        "Short",
        "Rarely",
        "No",
        "No",
        "No",
        "Long",
        "No",
    ],
    "Crawl": [
        "No",
        "Yes",
        "No",
        "No",
        "No",
        "No",
        "No",
        "Yes",
        "No",
        "Yes",
        "No",
        "No",
    ],
    "Class Label": [
        "Fish",
        "Animal",
        "Animal",
        "Animal",
        "Bird",
        "Bird",
        "Animal",
        "Animal",
        "Fish",
        "Fish",
        "Bird",
        "Bird",
    ],
}

df = pd.DataFrame(data)

classes = df["Class Label"].unique()

class_counts = Counter(df["Class Label"])
total_sampels = len(df)
prior_probs = {cls: class_counts[cls] / total_sampels for cls in classes}


def calc_likelihood(feature, value, cls):
    class_subset = df[df["Class Label"] == cls]
    count = len(class_subset[class_subset[feature] == value])

    return count / len(class_subset)


test_case = {"Swim": "Slow", "Fly": "Rarely", "Crawl": "No"}

final_probs = {}
for cls in classes:
    prob = 1
    for key, value in test_case.items():
        conditional_prob = calc_likelihood(key, value, cls)
        prob *= conditional_prob
    final_prob = prob * prior_probs[cls]
    final_probs[cls] = final_prob

max_key = max(final_probs, key=final_probs.get)
max_value = final_probs[max_key]

print(max_key, max_value)
```
